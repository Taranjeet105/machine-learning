```python
import matplotlib.pyplot as plt # need to be imported before seaborn
import pandas as pd
import seaborn as sns
import numpy as np
```

# Line Plot

```python
days=[1,2,3,4,5,6]
temp=[36,37,38,39.6,30,28]
temp_df=pd.DataFrame({"days":days,"temperature":temp})
sns.lineplot(x="days",y="temperature",data=temp_df)
plt.show()
```

```python
tips_df=sns.load_dataset("tips")
tips_df
```

```python
sns.lineplot(x="total_bill",y="tip",data=tips_df)
plt.show()
```

## categorical graph 

```python
tips_df=sns.load_dataset("tips")
sns.lineplot(x="total_bill",y="tip",data=tips_df,hue="sex") # customers are male or female so "hue" will categorize the graph
plt.show()
```

```python
tips_df=sns.load_dataset("tips")
sns.lineplot(x="total_bill",y="tip",data=tips_df,hue="sex",style="sex",palette="Accent") 
plt.show()
```

```python
tips_df=sns.load_dataset("tips")
plt.figure(figsize=(16,9))
sns.lineplot(x="total_bill",y="tip",data=tips_df,hue="sex",style="sex",palette="hot",markers=["o","<"]) 
plt.show()
```

## Adding labels and changing fontsize 

```python
tips_df=sns.load_dataset("tips")
plt.figure(figsize=(16,9))
sns.lineplot(x="total_bill",y="tip",data=tips_df,hue="sex",style="sex",palette="hot",markers=["o","<"]) 
plt.title("Tips dataset",fontsize=20)
plt.xlabel("Total_bill",fontsize=15)
plt.ylabel("Tip",fontsize=15)
plt.show()
```

## Adding style

```python
tips_df=sns.load_dataset("tips")
sns.set(style="darkgrid")
plt.figure(figsize=(16,9))
sns.lineplot(x="total_bill",y="tip",data=tips_df,hue="sex",style="sex",palette="hot",markers=["o","<"]) 
plt.title("Tips dataset",fontsize=20)
plt.xlabel("Total_bill",fontsize=15)
plt.ylabel("Tip",fontsize=15)
plt.show()
```

# Histogram

```python
tips_df=sns.load_dataset("tips")
tips_df
```

```python
sns.displot(tips_df["size"])
plt.show()
```

### KDE 

```python
sns.distplot(tips_df["size"],hist=False)
plt.show()
```

### Norm 

```python
from scipy.stats import norm
sns.distplot(tips_df["size"],hist=False,fit=norm,kde=False)
plt.legend(fontsize=20)
plt.show()
```

```python
sns.set()
sns.distplot(tips_df["total_bill"],label="Total bill")
plt.legend()
plt.show()
```

```python
tips_df.total_bill.sort_values()
```

```python
bins=[5,10,15,20,25,30,35,40,45,50,55]
sns.distplot(tips_df["total_bill"],bins=bins)
plt.xticks(bins)
plt.show()
```

### distplot is deprecated use histplot 

```python
bins=[5,10,15,20,25,30,35,40,45,50,55]
sns.histplot(tips_df["total_bill"],bins=bins)
plt.xticks(bins)
plt.show()
```

## hist_kws

```python
bins=[5,10,15,20,25,30,35,40,45,50,55]
sns.distplot(tips_df["total_bill"],bins=bins,hist_kws={
    'color':"silver",
    'edgecolor':"black",
    'alpha':0.9, # for darkness
    'linewidth':2
},kde=False)
plt.xticks(bins)
plt.show()
```

## kde_kws 

```python
bins=[5,10,15,20,25,30,35,40,45,50,55]
sns.distplot(tips_df["total_bill"],bins=bins,hist_kws={
    'color':"silver",
    'edgecolor':"black",
    'alpha':0.9, # for darkness
    'linewidth':2
},kde_kws={
    'color':"green",
    'alpha':0.9, # for darkness
    'linewidth':2
})
plt.xticks(bins)
plt.show()
```

# Barplot

```python
tips_df
```

```python
sns.barplot(x=tips_df.day,y=tips_df.total_bill)
plt.show()
```

```python
sns.barplot(x=tips_df.day,y=tips_df.total_bill, hue=tips_df.sex)
plt.title("Total bill pay by male and female",fontsize=20)
plt.show()
```

```python
sns.barplot(x='day',y='total_bill', hue='sex',data=tips_df)
plt.title("Total bill pay by male and female",fontsize=20)
plt.show()
```

```python
order=['Sun','Thur','Fri','Sat']
sns.barplot(x=tips_df.day,y=tips_df.total_bill, hue=tips_df.sex,order=order)
plt.title("Total bill pay by male and female",fontsize=20)
plt.show()
```

```python
hue_order=['Female','Male']
sns.barplot(x=tips_df.day,y=tips_df.total_bill, hue=tips_df.sex,order=order,hue_order=hue_order,estimator=np.max,ci=0) # estimator will me max value by default it is mean
plt.title("Total bill pay by male and female",fontsize=20)
plt.show()
```

```python
sns.barplot(y='day',x='total_bill', hue='sex',data=tips_df)
plt.title("Total bill pay by male and female",fontsize=20)
plt.show()
```

```python
sns.barplot(y='tip',x='total_bill', hue='sex',data=tips_df.head(10),orient='h',palette='hot',saturation=.1)
plt.title("Total bill pay by male and female",fontsize=20)
plt.show()
```

```python
sns.barplot(x='day',y='total_bill', hue='sex',data=tips_df,dodge=False) #dodge=False for overlapping barplots
plt.title("Total bill pay by male and female",fontsize=20)
plt.show()
```

```python
sns.set()
kwargs={'alpha':0.8,'linewidth':5,'edgecolor':'y','linestyle':':'}
sns.barplot(x='day',y='total_bill',data=tips_df,**kwargs)
plt.title("Total Bill VS Days",fontsize=20)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Total Bill ",fontsize=15)
plt.show()
```

### Assigning Axis to variable

```python
sns.set()
kwargs={'alpha':0.8,'linewidth':5,'edgecolor':'y','linestyle':':'}
ax=sns.barplot(x='day',y='total_bill',data=tips_df,**kwargs)
ax.set(title="Total bill with days",
      xlabel="Days",
      ylabel="Total Bill")
plt.show()
```
