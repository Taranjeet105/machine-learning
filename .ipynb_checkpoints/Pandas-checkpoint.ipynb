# Series

```python
import pandas as pd
```

```python
series1=pd.Series([1,2,3,'tarnjeet',3.4])
print(series1)
```

```python
series2=pd.Series([1,2,3,4],index=['a','b','c','d'],name='records1') ## naming and indexing
print(series2)
```

```python
series3=pd.Series([1,2,3,4],index=['a','b','c','d'],name='records2',dtype=float) ## changing datatype
print(series3)
```

```python
series4=pd.Series(0.23,index=[1,2,3,4,5]) ## series with scaler value
print(series4)
```

```python
series5=pd.Series({'a':1,'b':2,'c':3,'d':4}) ## series using dictionary
print(series5)
```

```python
series6=pd.Series([1,2,3,4,5,6,7,8,9]) 
print(series6)
```

```python
print(series6[0])
print()
print(series6[0:3])
print()
print(max(series6))
print()
print(min(series6))
print()
print(series6[series6>4])
```

```python
series6=series6*series6  ## mathematic operations
print(series6)
```

```python
series7=pd.Series([1,2,3])
series8=series6+series7
print(series8)
```

## dropna

```python
df5=pd.read_csv('../data.csv',nrows=5,usecols=[0,1,2,3,4],na_values=['not available']) 
df5
```

```python
df5.dropna(axis=0,how='any') # drops row which have NaN value/values in it. these are by default values
```

```python
df5.dropna(axis=1,how='any') # drops column which have NaN value/values in it
```

```python
df5.dropna(axis=1,how='all') # drops column if all values in that column are NaN
```

```python
df5.dropna(axis=0,how='all') # drops row if all values in that row are NaN
```

```python
df5.dropna(axis=0,how='any',thresh=4) # for columns to be there. it should atleast have 4 not null values
```

```python
df5
```

```python
df5.dropna(subset=['diagnosis']) # removes row where diagnosis is null
```

```python
# df5.dropna(inplace=true) modifies same dataframe. for inplace=false new dataframe is returned.
```

```python
df5.fillna(method='bfill') # takes downwards value
```

```python
df5.fillna(method='bfill',axis=1) #rotate dataframe 90 degree clockwise then same logic is apllied to fill using bfill in row
```

```python
df5.fillna(2,limit=1) # only 1 value will be fill with 2 for NaN
```

## Group by

```python
g1=df7.groupby('diagnosis') # df7.groupby('[a,b]').groups for mutliple columns
g1
```
